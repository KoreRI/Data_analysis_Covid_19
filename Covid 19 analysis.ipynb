```python
#IMPORT NECESSARY LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#pandas library helps to load the dataset file which is in csv format
df = pd.read_csv(r"D:\Urvi_PPT\JJTU\Python code\Covid-19\country_wise_latest.csv")
```

```python
#below line displays first 5 records of the dataset
df.head()
```

```python
#computes summary of statistics pertaining to the DataFrame columns.
df.describe()
```

```python
#check for NULL values if any in each column
df.isnull().sum()
```

```python
#summary of the DataFrame, including the index dtype and column dtypes, non-null values and memory usage
df.info()
```

```python
Region = df.groupby(by = 'WHO Region').sum()
Region.head()
Region[['Recovered','Active','Deaths','Confirmed']].plot(kind = 'bar')
plt.title('Recovered ,Active, Confirmed and Deaths Cases by Regions')
plt.xlabel('Region')
plt.ylabel('Count in millions')
```

```python
Top_10_country_confirmed = df.sort_values(["Confirmed"],ascending= False).head(10)
plt.figure(figsize = (10,5))
plt.style.use('dark_background')
sns.barplot(x = Top_10_country_confirmed['Confirmed'],y = Top_10_country_confirmed['Country/Region'],palette = 'Set1')
plt.title('Top 10 Countries with the Highest Confirmed Cases')
plt.xlabel('Number of Confirmed Cases (in millions)')
plt.ylabel('Country')
plt.show()
```

```python
Top_10_country_deaths = df.sort_values(["Deaths"],ascending= False).head(10)
plt.figure(figsize = (10,5))
plt.style.use('dark_background')
sns.barplot(x = Top_10_country_deaths['Deaths'],y = Top_10_country_deaths['Country/Region'],palette = 'RdBu')
plt.title('Top 10 Countries with the Highest Deaths')
plt.xlabel('Number of Deaths (in millions)')
plt.ylabel('Country')
plt.show()
```

```python
Top_10_country_recovered = df.sort_values(["Recovered"],ascending= False).head(10)
plt.figure(figsize = (10,5))
plt.style.use('dark_background')
sns.barplot(x = Top_10_country_recovered['Recovered'],y = Top_10_country_recovered['Country/Region'],palette = 'Set3')
plt.title('Top 10 Countries with the Highest Recovered Cases')
plt.xlabel('Number of Confirmed Cases (in millions)')
plt.ylabel('Country')
plt.show()
```

```python
Correlation_matrix = df[['Recovered','Active','Deaths','Confirmed']].corr()
Correlation_matrix
```

```python
sns.heatmap(Correlation_matrix,annot = True,cmap = 'BuPu',fmt = '.1f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
Region[['New cases','New deaths','New recovered']].plot(kind = 'line',rot=90)
plt.title(' New cases,New deaths and New recovered Cases by Regions')
plt.xlabel('Region')
plt.ylabel('Count of Cases')
```

```python
df1 = df.groupby(by= 'Country/Region').sum()
df1=df1.sort_values(['1 week change'],ascending=False).head(5) 
plt.pie(df1['1 week change'],labels = df1.index,
        autopct = '%1.1f%%',pctdistance = 0.8,startangle = 305,
        colors = plt.cm.Paired.colors,textprops = {'fontsize':10})
plt.title('% Distribution of 1 week change in covid cases(TOP 5 countires) ',fontsize = 15,weight = 'bold',loc = 'left')
plt.show()
```
